```python
import sqlite3
```

```python
# DB connection and Cursor Set Up
conn = sqlite3.connect('/home/jia/Desktop/Archiver Tool/database/archiver_database.db')
cursor = conn.cursor()
```

```python
# Table Creation USER TABLE PRE UPLOAD SCAN
# Table Creation RECENTLY ADDED
# Table Creation SAFETY CHECK
# Table Creation RENAME TABLE
# Table Creation TRANSFER HISTORY
# Table Creation REACTOR INVENTORY
# Table Creation USER HISTORY

cursor.executescript('''
CREATE TABLE IF NOT EXISTS preupload_scan (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    encoded_date DATETIME,
    complete_name TEXT,
    hash_value TEXT,
    format TEXT,
    file_size TEXT,
    duration TEXT,
    frame_rate TEXT,
    width TEXT,
    height TEXT,
    display_aspect_ratio TEXT,
    user_id TEXT,
    camera_id TEXT
    
               
);

CREATE TABLE IF NOT EXISTS recently_added (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    encoded_date DATETIME,
    complete_name TEXT,
    hash_value TEXT,
    datetime_upload DATETIME
    format TEXT,
    file_size TEXT,
    duration TEXT,
    frame_rate TEXT,
    width TEXT,
    height TEXT,
    display_aspect_ratio TEXT,
    user_id TEXT,
    camera_id TEXT
               
);

CREATE TABLE IF NOT EXISTS safety_check (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    encoded_date DATETIME,
    complete_name TEXT,
    hash_value TEXT,
    datetime_upload DATETIME,
    data_origin TEXT
    format TEXT,
    file_size TEXT,
    duration TEXT,
    frame_rate TEXT,
    width TEXT,
    height TEXT,
    display_aspect_ratio TEXT,
    user_id TEXT,
    camera_id TEXT
               
);         



CREATE TABLE IF NOT EXISTS rename_table (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    encoded_date DATETIME,
    complete_name TEXT,
    hash_value TEXT,
    datetime_upload DATETIME,
    new_name TEXT
    format TEXT,
    file_size TEXT,
    duration TEXT,
    frame_rate TEXT,
    width TEXT,
    height TEXT,
    display_aspect_ratio TEXT,
    user_id TEXT,
    camera_id TEXT
               
);

CREATE TABLE IF NOT EXISTS transfer_hist (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    encoded_date DATETIME,
    complete_name TEXT,
    hash_value TEXT,
    datetime_upload DATETIME,
    new_name TEXT
    format TEXT,
    file_size TEXT,
    duration TEXT,
    frame_rate TEXT,
    width TEXT,
    height TEXT,
    display_aspect_ratio TEXT,
    user_id TEXT,
    camera_id TEXT
               
);               

CREATE TABLE IF NOT EXISTS reactor_inv (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    encoded_date DATETIME,
    complete_name TEXT,
    hash_value TEXT,
    datetime_upload DATETIME,
    new_name TEXT
    format TEXT,
    file_size TEXT,
    duration TEXT,
    frame_rate TEXT,
    width TEXT,
    height TEXT,
    display_aspect_ratio TEXT,
    user_id TEXT,
    camera_id TEXT
               
);
           

CREATE TABLE IF NOT EXISTS user_hist (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    encoded_date DATETIME,
    complete_name TEXT,
    hash_value TEXT,
    datetime_upload DATETIME,
    new_name TEXT
    format TEXT,
    file_size TEXT,
    duration TEXT,
    frame_rate TEXT,
    width TEXT,
    height TEXT,
    display_aspect_ratio TEXT,
    user_id TEXT,
    camera_id TEXT
               
);

               

''')

conn.commit()
conn.close()
```
